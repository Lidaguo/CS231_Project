```python
import os, glob, pandas as pd, numpy as np

master_list = './data/allFrames/lists/shuffle_shotdetect_sampled_5_both_list.txt'

train_list = './data/allFrames/lists/shuffle_shotdetect_sampled_5_train_list.txt'
test_list = './data/allFrames/lists/shuffle_shotdetect_sampled_5_test_list.txt'

data_dir = 'data/optical_flow/stack_1_offset_30/'

def pathname_to_video_name(pathname):
    return '_'.join(os.path.basename(pathname).split('_')[:-1])
```

```python
test_videos = pd.read_csv(test_list, header = None, names = ['pathname', 'class_id'], delimiter= ' ')
test_video_list = list(test_videos.pathname.apply(lambda x: x.split('/')[0]).values)
test_video_list
```

```python
video_to_labels = pd.read_csv(master_list, delimiter= ' ', header=None, names = ['pathname', 'class_id'])
video_to_labels['video_name'] = video_to_labels.pathname.apply(lambda x: x.split('/')[0])
```

```python
vid_to_label = {k: v for k, v in zip(video_to_labels.video_name.values, video_to_labels.class_id.values)}
```

```python
opt_flow_list = []
for folder_id in range(1, 11):
    opt_flow_list.extend(glob.glob(os.path.join(data_dir, '%d'%folder_id, '*.jpg')))
len(opt_flow_list)
```

```python
x_list, y_list = [k for k in opt_flow_list if 'dX.jpg' in k], [k for k in opt_flow_list if 'dY.jpg' in k]
df = pd.DataFrame()
df['x'] = x_list
df['y'] = df.x.apply(lambda x: x[:-5] + 'Y.jpg')
df['exists'] = df.y.apply(lambda y: os.path.isfile(y))
df['video_name'] = df.x.apply(pathname_to_video_name)
df['class_label'] = df.video_name.apply(lambda x: vid_to_label[x])
```

```python
df['is_test']  = df.video_name.apply(lambda x: x in test_video_list)
```

```python
df.is_test.mean()
```

```python
path_header = '/afs/cs.stanford.edu/u/anenberg/scr/CS231N/'
df['x_fullpath'] = df.x.apply(lambda x: os.path.join(path_header, x))
df['y_fullpath'] = df.y.apply(lambda x: os.path.join(path_header, x))
```

```python
df.head(10)
```

```python
df_test = df[df.is_test]
df_train = df[df.is_test.apply(lambda x: not x)]

```

```python
lists_dir = '/afs/cs.stanford.edu/u/anenberg/scr/CS231N/data/allFrames/lists'

df_test[['x_fullpath', 'class_label']].to_csv(os.path.join(lists_dir, 'opt_flow_x_test.txt'), sep = ' ', index = False, header = False)
df_test[['y_fullpath', 'class_label']].to_csv(os.path.join(lists_dir, 'opt_flow_y_test.txt'), sep = ' ', index = False, header = False)

df_train[['x_fullpath', 'class_label']].to_csv(os.path.join(lists_dir, 'opt_flow_x_train.txt'), sep = ' ', index = False, header = False)
df_train[['y_fullpath', 'class_label']].to_csv(os.path.join(lists_dir, 'opt_flow_y_train.txt'), sep = ' ', index = False, header = False)
```
